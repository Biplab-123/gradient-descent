```python
#creating My own class  for SGDREGRESSor for multiple input coloumn 
```

```python
class SGDREGRESSOR():
    def __init__(self,learning_rate=0.01,epochs=100):
        self.intercept_=None
        self.coef_=None
        self.lr=learning_rate
        self.epochs=epochs
        
    def fit(self,x_train,y_train):
        #intialize the coef_
        self.intercept_ = 0
        self.coef_ = np.ones(x_train.shape[1])

        for i in range(self.epochs):
            for j in range(x_train.shape[0]):  # as we take one row at a time to calculate the 
                idx=np.random.randint(0,x_train.shape[0])   # creating the random number(index number) for the row in which the iteration will happen
                y_hat = np.dot(x_train[idx],self.coef_) + self.intercept_

                #For updating the intercept value for each randomly selected row 
                intercept_der = -2 * (y_train[idx] - y_hat)  # calculating the derivative using -2*(yi - y^) this formula for a single row value
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)

                #For updating the coeficient value for each randomly selected row 
                coef_der = -2 * np.dot((y_train[idx] - y_hat),x_train[idx])
                self.coef_ = self.coef_ - (self.lr * coef_der)
        print(self.intercept_ ,self.coef_)

    
    def predict(self,x_test):
        return np.dot(x_test,self.coef_) + self.intercept_ #y= mx+c for more than one number of coef and intercept
        
```

```python
#creating object
sgd = SGDREGRESSOR(learning_rate=0.01,epochs=50)
```

```python
start = time.time()
sgd.fit(x_train,y_train)
print("time taken to train",time.time() - start)
```

```python
y_pred = sgd.predict(x_test)
r2_score(y_test,y_pred)
```
